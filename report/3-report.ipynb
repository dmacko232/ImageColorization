# Report

In this project I tried to tackle the problem of Image Colorization.

During SOTA review I was surprised that there were not that many surveys that would easily answer the question I had. Still, I was able to summarize some of the information. I also discovered the issue that until recently there were no benchmark datasets intended for image colorization! Because many pictures from other datasets (ImageNET) can be sometimes hard to qualitatively evaluate I chose the Natural-Color Dataset with around 700 images -- since it should be straight forward how the color should look.

For the model I implemented ChromaGAN. This was a really painful experience and I actually wanted to quit and resort to segmentation. First, it was hard to define the architecture since it was not a simple one. Second, the pairing of discriminator and generator was complicated. Here I also had issues with gradients being None because of wrongly setting trainable parameter. This was the point when I wanted to quit. Also there was an issue where GradientPenaltyLoss I took from internet just didn't work at all in TF2 even when I tried many things. After lots of googling I discovered that there are some very complicated workarounds and the best approach is to rewrite the GAN without tf.keras functional API -- just plain vanilla tensorflow! I have also used in the past at work one Pose Estimation TF1 model library which THE AUTHORS were UNABLE to port into TF2 even after trying, so that discouraged me completely. As a result I decided to rather move back into TF1. Because I have a bit of experience with PyTorch, I will most likely in the past prefer PyTorch over TensorFlow where possible -- the code just seems much more readable, honestly.

For the training I used Google Collab. I used validation set of which I displayed sample images after each epoch. I experimented a bit with epochs and parameters but there was not much change. The training also seemed relatively stable -- maybe I attribute it to WGAN architecture. As far as the training, I only trained the model for 100 epochs -- I just didnt have enough time for more.

I found out that many metrics for colorization are not that good -- and that they can be confusing (sometimes optimizing for some metrics can lead to worse results -- for example too much brown color!). Thus, I decided to focus on qualitative evaluation. It was hard to pick a good baseline to compare it to. I found virtually no meaningful baselines that dont need any training. Thus I decided as a baseline to use segmetation architecture, I chose ColorfulColorization. Sometimes (overall ChromaGAN seems to have better performance) our model results were better and sometimes the segmentation architecture proved better. Still, both models seemed to usually have problems with green color. On the other hand the baseline also had issues where it sometimes colored something as green instead of red. Also there was a situation where segmentation architecture colored the white background! 

I guess the main experience of the project would be the model definition and debugging of such a complex network. In the future I would like to try out different image processing projects (maybe even the ones that were possible project topics) where I wouldnt need so much to create a model "from scratch" but I could focus more on comparing different models or just tweaking them (our I would use PyTorch more).
